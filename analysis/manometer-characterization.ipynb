# Questions
1. Can the instrument setup detect pressure changes?
2. Can the instrument setup detect the pressure changes generated by the venturi?  What is measurement?
3. How does the instrument setup (inc venturi) compare with reality?

# First Attempt
## Setup
1. Large manometer (1/4" PE) with one end open to the room
2. Other end connected through smallest silicon tube to the throat of the venturi tube
3. Fan connected to upstream port of the venturi tube
4. Fan voltage at 17, fan speed at 255
5. With thumb, plug and release downstream port of venturi tube
6. Measure pressure indicated on manometer

## Result
1. No change in reading
2. Try with small manometer: 3mmH2O movement on each side

## Conclusion
1. Slow leak in large manometer exceeds input airflow generated by pressure in venturi tube
2. Stiction losses in second manometer severely reduced measured pressure

# Second Attempt
## Setup
Same setup as before except as noted

## Measurements
1. Large manometer connected directly to fan: 9mmH2O
1. Connected to throat of venturi: 4mmH2O
    1. Water level did not change on its own.  Had to rock manometer and then let it settle to new reading
1. Switch manometer tube to 1/4" silicon.  tube is long enough to connect to venturi on its own
    1. Connect to throat: 5mmH20
    1. Connect to upstream: 9mmH20 (total)
1. Set fan speed to 128 i.e. half: 3mmH2o

# Conclusion
1. Can the instrument setup detect pressure changes? Yes, depending on manometer used and air tighness of connections
1. Can the instrument setup detect the pressure changes generated by the venturi? What is measurement?  Yes, 9mmH2O when fans are at full
1. How does the instrument setup (inc venturi) compare with reaty?  Uncertain.    Pressure measurement at venturi might be non-linear (based only on the two data points we have), suggesting it is measuring bernouli principle (pressure proportional to square of velocity).  The silicon tube meniscus is very difficult to read and not very well formed, making precise measurements more difficulty?

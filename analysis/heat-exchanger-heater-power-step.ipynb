```python
%matplotlib inline

import pandas as pd
import matplotlib.pyplot as plt
```

```python
def plot_temp(csv_file_name, title):
    data = pd.read_csv(csv_file_name)

    plt.title(title)
    plt.ylabel('Temperature [C]')
    plt.xlabel('Duty Cycle [/255]')
    plt.scatter(y=data["SENS0_TEMP"], x=data["HEATER1"], label="SENS0_TEMP")
    plt.scatter(y=data["SENS1_TEMP"], x=data["HEATER1"], label="SENS1_TEMP")
    plt.scatter(y=data["SENS2_TEMP"], x=data["HEATER1"], label="SENS2_TEMP")
    plt.scatter(y=data["SENS3_TEMP"], x=data["HEATER1"], label="SENS3_TEMP")
    plt.legend()
    plt.show()
```

```python
plot_temp("data/2024-09-19-heat-exchanger-heater-power-1.csv",
'''Heat Exchanger Port Temperatures vs Heater Duty Cycle
2024-09-19-heat-exchanger-heater-power-1.csv
Setup unchanged from Mike's house''')
```

```python
def plot_relative_temp(csv_file_name, title):
    data = pd.read_csv(csv_file_name)

    starting_ambient_temp = data["SENS2_TEMP"][0]
    data["SENS0_TEMP_CALIB"] = data["SENS0_TEMP"] - (data["SENS2_TEMP"] - starting_ambient_temp)
    data["SENS1_TEMP_CALIB"] = data["SENS1_TEMP"] - (data["SENS2_TEMP"] - starting_ambient_temp)
    data["SENS3_TEMP_CALIB"] = data["SENS3_TEMP"] - (data["SENS2_TEMP"] - starting_ambient_temp)
        
    plt.title(title)
    plt.ylabel('Temperature [C]')
    plt.xlabel('Duty Cycle [/255]')
    plt.scatter(y=data["SENS0_TEMP_CALIB"], x=data["HEATER1"], label="SENS0_TEMP_CALIB")
    plt.scatter(y=data["SENS1_TEMP_CALIB"], x=data["HEATER1"], label="SENS1_TEMP_CALIB")
    plt.scatter(y=data["SENS3_TEMP_CALIB"], x=data["HEATER1"], label="SENS3_TEMP_CALIB")
    plt.grid(True)
    plt.legend()
    plt.show()
    
plot_relative_temp("data/2024-09-19-heat-exchanger-heater-power-1.csv",
'''Heat Exchanger Relative (to T2) Port Temperatures vs Heater Duty Cycle
2024-09-19-heat-exchanger-heater-power-1.csv
Setup unchanged from Mike's house''')
```
